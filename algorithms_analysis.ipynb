<div class="entry-content"><header class="entry-header"><h1 class="entry-title" style="display:inline">Design and Analysis of Algorithms</h1><div class="page_content"><h2>What is meant by Algorithm Analysis?</h2><p>Algorithm analysis is an important part of computational complexity theory, which provides theoretical estimation for the required resources of an algorithm to solve a specific computational problem. Analysis of algorithms is the determination of the amount of time and space resources required to execute it.</p><h2>Why Analysis of Algorithms is important?</h2><ul><li>To predict the behavior of an algorithm without implementing it on a specific computer.</li><li>It is much more convenient to have simple measures for the efficiency of an algorithm than to implement the algorithm and test the efficiency every time a certain parameter in the underlying computer system changes.</li><li>It is impossible to predict the exact behavior of an algorithm. There are too many influencing factors.</li><li>The analysis is thus only an approximation; it is not perfect.</li><li>More importantly, by analyzing different algorithms, we can compare them to determine the best one for our purpose.</li></ul><h2>Types of Algorithm Analysis:</h2><ol><li>Best case</li><li>Worst case</li><li>Average case</li></ol></div></div></div></div></div>

<article class="content post-182475 post type-post status-publish format-standard hentry category-analysis tag-algorithms-analysis-of-algorithms" id="post-182475"><div style="position:relative"><div class="article-title"><h1>Analysis of Algorithms | Big-O analysis</h1></div>
<strong>Big-O Analysis of Algorithms</strong></p><p>We can express algorithmic complexity using the big-O notation. For a problem of size N:</p><div id="GFG_AD_gfg_mobile_336x280"></div><ul><li>A constant-time function/method is “order 1” : O(1)</li><li>A linear-time function/method is “order N” : O(N)</li><li>A quadratic-time function/method is “order N squared” : O(N 2 )</li></ul><p>Definition: Let g and f be functions from the set of natural numbers to itself. The function f is said to be O(g) (read big-oh of g), if there is a constant <strong>c &gt; 0</strong> and a natural number <strong>n<sub>0</sub></strong> such that f (n) ≤ cg(n) for all n &gt;= n<sub>0</sub> .</p><p>Note: O(g) is a set!</p><div style="min-height:250px;text-align:center;margin:20px 0" id="GFG_AD_Desktop_PostContent_300x250" data-google-query-id=""><div id="google_ads_iframe_/27823234/GFG_Desktop_PostContent_300x250_0__container__" style="border: 0pt none; width: 300px; height: 0px;"></div></div><p>Abuse of notation: f = O(g) does not mean f ∈ O(g).<br>The Big-O Asymptotic Notation gives us the Upper Bound Idea, mathematically described below:&nbsp;</p><blockquote><p>f(n) = O(g(n)) if there exists a positive integer n<sub>0</sub> and a positive constant c, such that f(n)≤c.g(n) ∀ n≥n<sub>0</sub>&nbsp;</p></blockquote><p>The general step wise procedure for Big-O runtime analysis is as follows:&nbsp;&nbsp;</p><ol><li>Figure out what the input is and what n represents.</li><li>Express the maximum number of operations, the algorithm performs in terms of n.</li><li>Eliminate all excluding the highest order terms.</li><li>Remove all the constant factors.</li></ol><p>Some of the useful properties of Big-O notation analysis are as follow:&nbsp;</p><blockquote><p> Constant Multiplication:&nbsp;<br>If f(n) = c.g(n), then O(f(n)) = O(g(n)) ; where c is a nonzero constant.&nbsp;<br> Polynomial Function:&nbsp;<br>If f(n) = a<sub>0</sub> + a<sub>1</sub>.n + a<sub>2</sub>.n<sup>2</sup> + —- + a<sub>m</sub>.n<sup>m</sup>, then O(f(n)) = O(n<sup>m</sup>).&nbsp;<br> Summation Function:&nbsp;<br>If f(n) = f<sub>1</sub>(n) + f<sub>2</sub>(n) + —- + f<sub>m</sub>(n) and f<sub>i</sub>(n)≤f<sub>i+1</sub>(n) ∀ i=1, 2, —-, m,&nbsp;<br>then O(f(n)) = O(max(f<sub>1</sub>(n), f<sub>2</sub>(n), —-, f<sub>m</sub>(n))).&nbsp;<br> Logarithmic Function:&nbsp;<br>If f(n) = log<sub>a</sub>n and g(n)=log<sub>b</sub>n, then O(f(n))=O(g(n))&nbsp;<br>; all log functions grow in the same manner in terms of Big-O.</p></blockquote><p>&nbsp;Basically, this asymptotic notation is used to measure and compare the worst-case scenarios of algorithms theoretically. For any algorithm, the Big-O analysis should be straightforward as long as we correctly identify the operations that are dependent on n, the input size.&nbsp;<br><strong>Runtime Analysis of Algorithms</strong></p><p>In general cases, we mainly used to measure and compare the worst-case theoretical running time complexities of algorithms for the performance analysis.&nbsp;<br>The fastest possible running time for any algorithm is O(1), commonly referred to as <i>Constant Running Time</i>. In this case, the algorithm always takes the same amount of time to execute, regardless of the input size. This is the ideal runtime for an algorithm, but it’s rarely achievable.&nbsp;<br>In actual cases, the performance (Runtime) of an algorithm depends on n, that is the size of the input or the number of operations is required for each input item.&nbsp;<br>The algorithms can be classified as follows from the best-to-worst performance (Running Time Complexity):&nbsp;</p><blockquote><p> A logarithmic algorithm – O(logn)&nbsp;<br>Runtime grows logarithmically in proportion to n.&nbsp;<br> A linear algorithm – O(n)&nbsp;<br>Runtime grows directly in proportion to n.&nbsp;<br> A superlinear algorithm – O(nlogn)&nbsp;<br>Runtime grows in proportion to n.&nbsp;<br> A polynomial algorithm – O(n<sup>c</sup>)&nbsp;<br>Runtime grows quicker than previous all based on n.&nbsp;<br> A exponential algorithm – O(c<sup>n</sup>)&nbsp;<br>Runtime grows even faster than polynomial algorithm based on n.&nbsp;<br> A factorial algorithm – O(n!)&nbsp;<br>Runtime grows the fastest and becomes quickly unusable for even&nbsp;<br>small values of n.&nbsp;&nbsp;</p></blockquote><p>Where, n is the input size and c is a positive constant.&nbsp;<br>&nbsp;</p><div style="min-height:90px;text-align:center;margin:20px 0" id="GFG_AD_Desktop_MTF_Postcontent_728x90"></div><p><img src="https://media.geeksforgeeks.org/wp-content/cdn-uploads/mypic.png" alt="asymtotic-analysis"></p><p><strong>Algorithmic Examples of Runtime Analysis</strong>:&nbsp;<br>Some of the examples of all those types of algorithms (in worst-case scenarios) are mentioned below:&nbsp;</p><blockquote><p> Logarithmic algorithm – O(logn) – Binary Search.&nbsp;<br> Linear algorithm – O(n) – Linear Search.&nbsp;<br> Superlinear algorithm – O(nlogn) – Heap Sort, Merge Sort.&nbsp;<br> Polynomial algorithm – O(n^c) – Strassen’s Matrix Multiplication, Bubble Sort, Selection Sort, Insertion Sort, Bucket Sort.&nbsp;<br> Exponential algorithm – O(c^n) – Tower of Hanoi.&nbsp;<br> Factorial algorithm – O(n!) – Determinant Expansion by Minors, Brute force Search algorithm for Traveling Salesman Problem.&nbsp;</p></blockquote><p><strong>Mathematical Examples of Runtime Analysis</strong>:&nbsp;<br>The performances (Runtimes) of different orders of algorithms separate rapidly as n (the input size) gets larger. Let’s consider the mathematical example:&nbsp;&nbsp;</p><pre>If n = 10,                  If n=20,
    log(10) = 1;                log(20) = 2.996;
    10 = 10;                    20 = 20;
    10log(10)=10;               20log(20)=59.9;
    10<sup>2</sup>=100;                    20<sup>2</sup>=400;
    2<sup>10</sup>=1024;                    2<sup>20</sup>=1048576;
    10!=3628800;                20!=2.432902e+18<sup>18</sup>;</pre><p><strong>Memory Footprint Analysis of Algorithms</strong></p><p>For performance analysis of an algorithm, runtime measurement is not only relevant metric but also we need to consider the memory usage amount of the program. This is referred to as the Memory Footprint of the algorithm, shortly known as Space Complexity.&nbsp;<br>Here also, we need to measure and compare the worst case theoretical space complexities of algorithms for the performance analysis.&nbsp;<br>It basically depends on two major aspects described below:&nbsp;</p><ul><li>Firstly, the implementation of the program is responsible for memory usage. For example, we can assume that recursive implementation always reserves more memory than the corresponding iterative implementation of a particular problem.</li><li>And the other one is n, the input size or the amount of storage required for each item. For example, a simple algorithm with a high amount of input size can consume more memory than a complex algorithm with less amount of input size.</li></ul><p>Algorithmic Examples of Memory Footprint Analysis: The algorithms with examples are classified from the best-to-worst performance (Space Complexity) based on the worst-case scenarios are mentioned below:&nbsp;&nbsp;</p><pre> Ideal algorithm - O(1) - Linear Search, Binary Search,
    Bubble Sort, Selection Sort, Insertion Sort, Heap Sort, Shell Sort.
 Logarithmic algorithm - O(log n) - Merge Sort.
 Linear algorithm - O(n) - Quick Sort.
 Sub-linear algorithm - O(n+k) - Radix Sort.</pre><p><strong>Space-Time Tradeoff and Efficiency</strong></p><p>There is usually a trade-off between optimal memory use and runtime performance.&nbsp;<br>In general for an algorithm, space efficiency and time efficiency reach at two opposite ends and each point in between them has a certain time and space efficiency. So, the more time efficiency you have, the less space efficiency you have and vice versa.&nbsp;<br>For example, Mergesort algorithm is exceedingly fast but requires a lot of space to do the operations. On the other side, Bubble Sort is exceedingly slow but requires the minimum space.&nbsp;<br>At the end of this topic, we can conclude that finding an algorithm that works in less running time and also having less requirement of memory space, can make a huge difference in how well an algorithm performs.</p><p>Example of Big-oh noatation:</p><div class="responsive-tabs-wrapper"><div
